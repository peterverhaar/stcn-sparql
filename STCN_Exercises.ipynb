## Question 1. 

Try the find the title of the book with the following identifier:

&lt;http://data.bibliotheken.nl/id/nbt/p216567742&gt;

The title is provided as an object for the `schema:name` predicate. 

Use the SPARQL Query below as a basis. 

```
prefix schema: <http://schema.org/>
prefix rdfs:  <http://www.w3.org/2000/01/rdf-schema#>
prefix kb: <http://data.bibliotheken.nl/def#>
SELECT * 
WHERE {
  <http://data.bibliotheken.nl/id/nbt/p216567742> schema:inLanguage ?language .
}
```

## Question 3.

If you want to see the values for multiple predicates of one specific book, defined by a URI, it can be useful, firstly, to provide this URI as the value of the `schema:mainEntity` property, as in the following query.  

```
PREFIX schema: <http://schema.org/>
PREFIX kb: <http://data.bibliotheken.nl/def#>

SELECT *
WHERE
{ 

?book schema:mainEntityOfPage ?mainentity . 
?mainentity schema:mainEntity <http://data.bibliotheken.nl/id/nbt/p412692457> . 

?book schema:name ?title .
}
```

The query, at this point, only returns the title. Could you add some code to ensure that the query also returns the language and the collation formula? These values are given after `schema:inLanguage` and `kb:hasCollationalFormula`.


## Question 4.

The year of publication of the book with identifier &lt;http://data.bibliotheken.nl/id/nbt/p412692457&gt; can be found using the following query. 


```
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
PREFIX schema: <http://schema.org/>
PREFIX kb: <http://data.bibliotheken.nl/def#>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

SELECT *
WHERE
{ 

?book schema:mainEntityOfPage ?mainentity . 
?mainentity schema:mainEntity <http://data.bibliotheken.nl/id/nbt/p412692457> .

?book schema:name ?title .
?book schema:publication ?publ .
?publ schema:startDate ?publ_year . 

}
```

Using this query as a basis, cab you create a list of all the titles printed in the year 1710?


## Question 5.

The query below returns all the works by Erasmus. 


```
PREFIX schema: <http://schema.org/>
PREFIX kb: <http://data.bibliotheken.nl/def#>

SELECT * WHERE {
?book schema:mainEntityOfPage ?mainEntity .
?mainEntity schema:isPartOf <http://data.bibliotheken.nl/id/dataset/stcn>  . 

?book schema:author ?author_node . 
?author_node schema:author <http://data.bibliotheken.nl/id/thes/p068551959> .

?book schema:name ?title . 
}
```

Can you find all the works in the STCN written by John Milton? The URI of Milton is 

&lt;http://data.bibliotheken.nl/id/thes/p068421443&gt;
